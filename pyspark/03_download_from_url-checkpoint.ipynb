```python
import requests
import os
import itertools
import zipfile
from pyspark import SparkFiles
from pyspark.sql import SparkSession # import library for sessions
```

```python
# Start spark session
spark = ( 
 SparkSession
 .builder
 .master("local[*]") # machine core usage configuration
 .appName('spark_dataframe_api') # session name
 .getOrCreate()
)
```

```python
def download_file(url, dest_folder="."):
    """Downloads a file from a URL to a local folder using the requests library."""
    if not os.path.exists(dest_folder):
        os.makedirs(dest_folder)
    
    filename = url.split('/')[-1]
    file_path = os.path.join(dest_folder, filename)
    
    try:
        with requests.get(url, stream=True) as r:
            r.raise_for_status() # Raise an exception for bad status codes
            with open(file_path, 'wb') as f:
                for chunk in r.iter_content(chunk_size=8192):
                    f.write(chunk)
        print(f"Downloaded {filename} to {os.path.abspath(file_path)}")
        return file_path
    except requests.exceptions.RequestException as e:
        print(f"Download failed: {e}")
        return None
```

```python
# defining variables to download files
path = "https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/arquivos/shpc/dsas/ca/ca-"
year = range(2004,2026)
quarter = [1,2]
year_quarter = list(itertools.product(year, quarter)) # Generate combinations as a list of tuples
local_download_path = "./temp_downloads"
extract_path = "./dados"
```

```python
# Download files and unzip it
for year, quarter in year_quarter:
    url = f"{path}{year}-0{quarter}.zip"
    # 1. Download the file to the driver's local disk
    downloaded_file = download_file(url, local_download_path)
    
    if downloaded_file:
        # 2. Add the locally downloaded file to the SparkContext, which distributes it to workers
        spark.sparkContext.addFile(downloaded_file)
        
        # 3. Access the file in a distributed job using SparkFiles.get()
        file_location = SparkFiles.get(os.path.basename(downloaded_file))
        print(f"File distributed to workers at: {file_location}")

        # 4. unzip the file
        with zipfile.ZipFile(file_location, 'r') as zip_ref:
            zip_ref.extractall(extract_path)
        print(f"File unzipped")
```
